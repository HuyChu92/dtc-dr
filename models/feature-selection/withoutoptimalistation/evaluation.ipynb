```python
import json
import pandas as pd
```

```python
# Open and read the JSON file
with open(r'D:\dtc-dr\models\feature-selection\combined_results_23012024.json', 'r') as file:
    # Parse the JSON content
    result = json.load(file)

decision_tree = {}
for key, value in result.items():
    decision_tree[key] = value["decision_tree"]
    decision_tree_train = [round(value[0], 2) for key, value in decision_tree.items()]
    decision_tree_test = [round(value[1], 2) for key, value in decision_tree.items()]

with open(r'D:\dtc-dr\models\feature-selection\mlp\cv_scores_mlp.json', 'r') as file:
    # Parse the JSON content
    mlp = json.load(file)
    mlp_train = [round(value["r2_scores_train"], 2) for key, value in mlp.items()]
    mlp_test = [round(value["r2_scores_test"], 2) for key, value in mlp.items()]

with open(r'D:\dtc-dr\models\feature-selection\linearregression\cv\cv_scores_linearregression.json', 'r') as file:
    # Parse the JSON content
    linearregression = json.load(file)
    linearregression_train = [round(value["r2_scores_train"], 2) for key, value in linearregression.items()]
    linearregression_test = [round(value["r2_scores_test"], 2) for key, value in linearregression.items()]

with open(r'D:\dtc-dr\models\feature-selection\randomforrest\Random_Forest_tree_best_models.json', 'r') as file:
    # Parse the JSON content
    random_forrest = json.load(file)
    random_forrest_train = [round(value["r2_scores_train"], 2) for key, value in random_forrest.items()]
    random_forrest_test = [round(value["r2_scores_test"], 2) for key, value in random_forrest.items()]

features = [key for key, value in result.items()]
random_forrest
```

```python

data = {
    'Features': features,
    'Decision Tree Train': decision_tree_train,
    'Decision Tree Test': decision_tree_test,
    'MLP Train': mlp_train,
    'MLP Test': mlp_test,
    'Linear Regression Train': linearregression_train,
    'Linear Regression Test': linearregression_test,
    'Random Forest Train': random_forrest_train,
    'Random Forest Test': random_forrest_test
}

df = pd.DataFrame(data)

# Set 'Features' column as index
df.set_index('Features', inplace=True)

# Display the DataFrame
df
df.to_excel("evaluation_regression_models.xlsx")
```
